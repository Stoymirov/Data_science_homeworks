```python
data_billboard = pd.read_csv('billboard.csv')
```

```python
data_billboard[['wk1','wk2']].dtypes
```

```python
data_billboard = data_billboard.melt(id_vars=['year','artist','track','time','date.entered'],var_name='week',value_name='position')
```

```python
data_billboard.week.value_counts()
```

```python
data_billboard.week = data_billboard.week.str.slice(2).astype(int)
```

```python
data_billboard
```

```python
data_billboard['date.entered'] = pd.to_datetime(data_billboard['date.entered'])
```

```python
data_billboard=data_billboard.dropna()
```

```python
data_billboard['date.entered'] + pd.Timedelta(f'{} days')
```

```python
(data_billboard.week -1) * 7
```

```python
data_billboard.pivot_table(index = 'track',
                           columns='artist',
                           values = 'date.entered',
                           aggfunc='count').to_csv('check.csv')
```
